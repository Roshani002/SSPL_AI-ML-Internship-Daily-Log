## Oversampling and Undersampling

### 1. Random Under-Sampling

Here in random under sampling `.sample()` returns a new dataframe, so we have to use new df.

```python
class_not_fraud, class_fraud = df['Class'].value_counts()

fraud = df[df['Class']==1]
not_fraud = df[df['Class']==0]

print("Fraud transcations", fraud.shape)
print("Non Fraud transcations", not_fraud.shape)
```

```python
not_fraud_under = not_fraud.sample(class_fraud)

undersample_df = pd.concat([not_fraud_under, fraud], axis=0)

print("total class of fraud and not fraud:", undersample_df['Class'].value_counts())
undersample_df['Class'].value_counts().plot(kind='bar', title=' count (target)')
```

### 2. Over-Sampling using imblearn

```python
x = df.drop('Class', axis=1)
y = df['Class']
```

```python
# spliting the dataset in testing and training sets

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42)

x_ros, y_ros = ros.fit_resample(x, y)

print('Original dataset shape', Counter(y))
print('Resample dataset shape Counter', Counter(y_ros))
```

### 3. Under-Sampling: Tomek Links

Tomek links are pairs of very close instances but of opposite classes. Removing the instances of the majority class of each pair increases the space between the two classes, facilitating the classification process.

Tomek’s link exists if the two samples are the nearest neighbors of each other.

![alt text](image.png)

```python
from imblearn.under_sampling import TomekLinks

tl = TomekLinks()
# tl = RandomOverSampler(sampling_strategy='not minority')

x_tl, y_tl = tl.fit_resample(x,y)

print('Original dataset shape', Counter(y))
print('Resample dataset shape', Counter(y_tl))
```

### 4. Synthetic Minority Oversampling Technique (SMOTE)

SMOTE (Synthetic Minority Oversampling Technique in machine learning) works by randomly picking a point from the minority class and computing the k-nearest neighbors for this point. The synthetic points are added between the chosen point and its neighbors.

![alt text](image-1.png)

SMOTE algorithm works in 4 simple steps:

1. Choose a minority class as the input vector.
2. Find its k nearest neighbors (k_neighbors is specified as an argument in the SMOTE() function).
3. Choose one of these neighbors and place a synthetic point anywhere on the line joining the point under consideration and its chosen neighbor.
4. Repeat the steps until the data is balanced.

```python
# import library
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

# fit predictor and target variable
x_smote, y_smote = smote.fit_resample(x, y)

print('Original dataset shape', Counter(y))
print('Resample dataset shape', Counter(y_smote))
```

### 5. NearMiss

NearMiss is an under-sampling technique. Instead of resampling the Minority class, using a distance will make the majority class equal to the minority class.

```python
from imblearn.under_sampling import NearMiss

nm = NearMiss()

x_nm, y_nm = nm.fit_resample(x, y)

print('Original dataset shape:', Counter(y))
print('Resample dataset shape:', Counter(y_nm))
```
